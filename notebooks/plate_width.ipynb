```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
DATA_PATH = 'train_data_old/single'
```

```python
plate = pd.DataFrame([], columns=['path', 'gTruth', 'length'])

for file in os.listdir(DATA_PATH):
    
    filename, file_extension = os.path.splitext(file)
    gt = filename.split('_')[-1]
    
    plate = plate.append({'path': file,
                          'gTruth': gt,
                          'length': len(gt)},
                           ignore_index=True)
```

```python
plate.length.value_counts()
```

```python
for i, row in plate.loc[plate.length==3][2:4].iterrows():
    img = Image.open(os.path.join(DATA_PATH, row['path']))
    img = img.resize((44,32))
    plt.imshow(img)
    plt.show()
```

```python
for i, row in plate.loc[plate.length==7][:2].iterrows():
    img = Image.open(os.path.join(DATA_PATH, row['path']))
    img = img.resize((108,32))
    plt.imshow(img)
    plt.show()
```

```python
for i, row in plate.loc[plate.length==6][:2].iterrows():
    img = Image.open(os.path.join(DATA_PATH, row['path']))
    img = img.resize((92,32))
    plt.imshow(img)
    plt.show()
```
